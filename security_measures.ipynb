```python
import pandas as pd
```

```python
DATA_FILE = 'data/gen/cleaned.csv'
```

```python
df = pd.read_csv(DATA_FILE)
df.head()
```

```python
# If exist remove the 'MEDIDAS_SEGURIDAD' column
if 'MEDIDAS_SEGURIDAD' in df.columns:
    df.drop('MEDIDAS_SEGURIDAD', axis=1, inplace=True)
```

```python
def determine_security_measure(row):
    if row['VIC_EDAD'] < 18:
        return 'Dto. 27-20033'

    if row['VIC_SEXO'] == 'Mujeres' and row['AGR_SEXO'] == 'Mujeres':
        return 'Dto. 97-96'

    if row['VIC_SEXO'] == 'Mujeres' and row['AGR_SEXO'] == 'Hombres':
        return 'Dto. 22-2008'

    if row['VIC_SEXO'] == 'Hombres':
        return 'Dto. 97-96'

    return None
```

```python
df['MEDIDAS_SEGURIDAD'] = df.apply(determine_security_measure, axis=1)
```

```python
print(df['MEDIDAS_SEGURIDAD'].value_counts())
```

```python
# Save the cleaned data
df.to_csv('data/gen/cleaned.csv', index=False)
```
